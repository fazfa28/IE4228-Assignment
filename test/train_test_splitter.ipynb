```python
import os
import shutil
import random
from pathlib import Path

# ── CONFIG ──────────────────────────────────────────────
SOURCE_DIR   = "../Images/After"          # root folder with 7 person subdirs
OUTPUT_DIR   = "../datasets/train_2"   # where train/ and test/ will be created
TRAIN_RATIO  = 0.8              # 80% train, 20% test
SEED         = 42               # for reproducibility
IMAGE_EXTS   = {".jpg", ".jpeg", ".png", ".bmp", ".gif", ".webp", ".tiff"}
# ────────────────────────────────────────────────────────

def split_dataset(source_dir, output_dir, train_ratio, seed):
    random.seed(seed)
    source_path = Path(source_dir)
    output_path = Path(output_dir)

    person_dirs = [d for d in source_path.iterdir() if d.is_dir()]
    if not person_dirs:
        print(f"No subdirectories found in '{source_dir}'")
        return

    print(f"Found {len(person_dirs)} person(s): {[d.name for d in person_dirs]}\n")

    for person_dir in sorted(person_dirs):
        images = [
            f for f in person_dir.iterdir()
            if f.is_file() and f.suffix.lower() in IMAGE_EXTS
        ]

        if not images:
            print(f"  [{person_dir.name}] No images found, skipping.")
            continue

        random.shuffle(images)
        split_idx  = max(1, int(len(images) * train_ratio))
        train_imgs = images[:split_idx]
        test_imgs  = images[split_idx:]

        for subset, files in [("train", train_imgs), ("test", test_imgs)]:
            dest = output_path / subset / person_dir.name
            dest.mkdir(parents=True, exist_ok=True)
            for f in files:
                shutil.move(str(f), dest / f.name)

        print(
            f"  [{person_dir.name}] {len(images)} images → "
            f"{len(train_imgs)} train / {len(test_imgs)} test"
        )

    print(f"\nDone! Output saved to: {output_path.resolve()}")

if __name__ == "__main__":
    split_dataset(SOURCE_DIR, OUTPUT_DIR, TRAIN_RATIO, SEED)
```
